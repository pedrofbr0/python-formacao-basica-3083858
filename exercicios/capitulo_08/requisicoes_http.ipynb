- Importe as bibliotecas requests e csv


```python
import requests, csv
```

- Faça uma requisição do tipo 'get' para o link 'http://dados.recife.pe.gov.br/dataset/2676fd74-8b40-4248-a1da-0dc45f176b7b/resource/7c005351-2d14-46cf-b2f0-6622a5553b30/download/ocorrencias2022.csv'


```python
link = 'http://dados.recife.pe.gov.br/dataset/2676fd74-8b40-4248-a1da-0dc45f176b7b/resource/7c005351-2d14-46cf-b2f0-6622a5553b30/download/ocorrencias2022.csv'

dados = requests.get(link)
```

- Imprima na tela o código com o status da requisição- Imprima na tela os metadados do arquivo


```python
print(dados.status_code)

dados.headers
```

- Salve os dados em um arquivo csv, na pasta 'dados'


```python
with open('/workspaces/python-formacao-basica-3083858/dados/samu_2022.csv', 'wb') as dados_csv:
    dados_csv.write(dados.content)
```
